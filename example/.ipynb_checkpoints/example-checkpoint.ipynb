# 1. Load data

```julia
# load curve
pc = CSV.read("random_curve.tsv", DataFrame)
pc = Array(pc[!,["x", "y", "z"]])

# load barcode and generators 
data = JSON.parsefile("PH_minimal.json")
curve_barcode = data["barcode"]
# convert format of barcode
curve_barcode = hcat(curve_barcode[1], curve_barcode[2])
min_rationals_generators = data["generators_v"]
generators = data["representatives"];
```

# 2. Plot curve

```julia
# plot curve
plot_curve(pc)
```

# 3. Plot persistence diagram

```julia
# plot barcode
p = plot_PD(curve_barcode)
```

# 4. Plot generators

(a) Plot the length-minimizing generators (with rational coefficients).  
(b) Plot the final generator after the jump-minimization step.

```julia
# select a class
i = 30
plot_PD(curve_barcode, highlight = [i]; markersize = 10)
```

```julia
# plot minimal generator
p1 = PH_minimal.plot_cycle_3D(pc, min_rationals_generators[string(i)], title = "minimal generator over rational")


plot(p1)
```

```julia
p2 = PH_minimal.plot_cycle_3D(pc, generators[i], title ="jump minimized")
plot(p2)
```
